```python
# math
from numpy import *
from scipy import interpolate

# astro & healpix
import healpy as hp
import pyfits as pf

# utilities
import glob
import os

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='dark', context='poster', color_codes=True)

# tdpy
import tdpy_util
```

```python
def diffuse_model(ener, nside=256, vfdm=6, cart=False):                    
    
    path = os.environ["PNTS_TRAN_DATA_PATH"]
    if cart:
        path += '/fdfm/diffcart/'
    else:
        path += '/fdfm/diffheal/'

    if vfdm == 2:
        path += 'gll_iem_v02.fit'
    if vfdm == 3:
        path += 'gll_iem_v02_P6_V11_DIFFUSE.fit'
    if vfdm == 4:
        path += 'gal_2yearp7v6_v0.fits'
    if vfdm == 5:
        path += 'gll_iem_v05.fit'
    if vfdm == 6:
        path += 'gll_iem_v05_rev1.fit'
    if vfdm == 7:
        path += 'gll_iem_v06.fits'
   
    if cart:
        fluxinpt = pf.getdata(path, 0)
        nenerinpt = fluxinpt.shape[0]
        nbghpinpt = fluxinpt.shape[1]
        nlghpinpt = fluxinpt.shape[2]
        fluxinpt = fluxinpt.flatten()

        minmlghpinpt = -180.
        maxmlghpinpt = 180.
        minmbghpinpt = -90.
        maxmbghpinpt = 90.
        enerinpt = array(pf.getdata(path, 1).tolist()).flatten() * 1e-3
        lghpinpt = linspace(minmlghpinpt, maxmlghpinpt, nlghpinpt)
        bghpinpt = linspace(minmbghpinpt, maxmbghpinpt, nbghpinpt)
        
        # input grid
        gridinpttemp = meshgrid(enerinpt, lghpinpt, bghpinpt)
        gridinpt = zeros((nenerinpt * nlghpinpt * nbghpinpt, 3))
        gridinpt[:,0] = gridinpttemp[0].flatten()
        gridinpt[:,1] = gridinpttemp[1].flatten()
        gridinpt[:,2] = gridinpttemp[2].flatten()
        
        # output grid
        nener = ener.size
        lghp, bghp, nside, npixl, apix = tdpy_util.get_heal(nside)
        gridoutptemp = meshgrid(ener, arange(npixl, dtype=int))
        gridoutp = zeros((nener * npixl, 3))
        gridoutp[:,0] = gridoutptemp[0].flatten()
        gridoutp[:,1] = lghp[gridoutptemp[1].flatten()]
        gridoutp[:,2] = bghp[gridoutptemp[1].flatten()]

        # interpolate
        fdfmflux = interpolate.griddata(gridinpt, fluxinpt, gridoutp)
        
    else:
        fluxfdfm = array(pf.getdata(path, 0).tolist()) # [1/cm^2/s/sr/GeV]
        enerfdfm = array(pf.getdata(path, 1).tolist()).flatten() * 1e-3 # [GeV]
        fluxinpt = interpolate.interp1d(enerfdfm, fluxfdfm, axis=0)
        fdfmflux = fluxinpt(ener)
        
    return fdfmflux

```

```python
def flist(version, nside, evclass, evtype):

    tag = os.environ["PNTS_TRAN_DATA_PATH"] + "/" + version + "/flux_*_%4.4d_evc%3.3d_evt%2.2d.fits" % (nside, evclass, evtype)
    flist = sorted(glob.glob(tag))

    return flist

def flist_legacy(version, nopsc=False):

    tag = os.environ["FERMI_DATA"] + "allsky/" + version + "/specbin/fermi-allsky-*-fwhm000-0256"
    if nopsc:
        tag += "-nopsc"
    tag += ".fits"
    flist = sorted(glob.glob(tag))

    return flist
```
